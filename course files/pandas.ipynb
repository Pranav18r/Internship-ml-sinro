#Sorting Data

```python
import pandas as pd

data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Age": [23, 25, 22, 24, 28],
    "Marks": [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)
df
```

```python
# Sort by Marks

df.sort_values("Marks")
```

```python
# Descending order
df.sort_values("Marks", ascending=False)
```

```python
# Multiple columns

df.sort_values(["Marks", "Age"])
```

#GroupBy

```python
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Department": ["HR", "IT", "HR", "IT", "HR"],
    "Salary": [30000, 50000, 35000, 60000, 40000]
}

df = pd.DataFrame(data)
df
```

```python
# Group by Department

df.groupby("Department")
```

```python
# Aggregation

# Average salary per department

df.groupby("Department")["Salary"].mean()
```

```python
# Multiple aggregations

df.groupby("Department")["Salary"].agg(["mean", "sum", "max"])
```

```python
# Count employees per department

df.groupby("Department")["Name"].count()
```

#Mini Project: Student Performance Analysis

```python
import pandas as pd

data = {
    "Name": ["A", "B", "C", "D", "E", "F", "G", "H"],
    "Department": ["CS", "IT", "CS", "IT", "EC", "CS", "EC", "IT"],
    "Age": [20, 21, 19, 22, 20, 23, 21, 20],
    "Marks": [85, 78, 92, 70, 88, 95, 60, 82]
}

df = pd.DataFrame(data)
```

**Basic Exploration**
1. Show first 5 rows
2. Show last 3 rows
3. Find:
  *   number of rows & columns
  *   column names
  *   data types

**Selection & Filtering**

4. Select only "Name" and "Marks" columns
5. Show students who scored more than 80
6. Show students with:
  * Marks > 80
  * Age < 21
7. Show students from CS department

**Sorting**

8. Sort students by Marks (ascending)
9. Sort students by Marks (descending)
10. Sort by Department, then Marks

**GroupBy + Aggregation**

11. Find average marks per department
12. Find total marks per department
13. Find maximum marks per department
14. Count number of students in each department

```python
# Part 1
df.head()
df.tail(3)
df.shape
df.columns
df.info()

# Part 2
df[["Name", "Marks"]]
df[df["Marks"] > 80]
df[(df["Marks"] > 80) & (df["Age"] < 21)]
df[df["Department"] == "CS"]

# Part 3
df.sort_values("Marks")
df.sort_values("Marks", ascending=False)
df.sort_values(["Department", "Marks"])

# Part 4
df.groupby("Department")["Marks"].mean()
df.groupby("Department")["Marks"].sum()
df.groupby("Department")["Marks"].max()
df.groupby("Department")["Name"].count()
```
