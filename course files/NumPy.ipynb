#Basic NumPy functions

```python
arr = np.array([[10,20,30],[40,50,60]])
```

```python
#mean

print(np.mean(arr))

#columnwise mean

print(np.mean(arr,axis=0))

#rowwise mean

print(np.mean(arr,axis=1))
```

```python
#max

print(np.max(arr))

#columnwise max

print(np.max(arr,axis=0))

#rowwise max

print(np.max(arr,axis=1))
```

```python
#min

print(np.min(arr))

#columnwise min

print(np.min(arr,axis=0))

#rowwise min

print(np.min(arr,axis=1))
```

```python
#sum

print(np.sum(arr))

#columwise

print(np.sum(arr,axis=0))

#rowwise

print(np.sum(arr,axis=1))
```

```python
#median

print(np.median(arr))

#columnwise

print(np.median(arr,axis=0))

#rowwise

print(np.median(arr,axis=1))
```

```python
#standard deviation

# This treats all values as a single list:
# [10,20,30,40,50,60]

# 1.Mean
# (10+20+30+40+50+60)/6=210/6=35
# 2.Squared differences
# (10 - 35)² = 625
# (20 - 35)² = 225
# (30 - 35)² = 25
# (40 - 35)² = 25
# (50 - 35)² = 225
# (60 - 35)² = 625
# Sum = 1750
# 3.Variance
# 1750/6≈291.67
# 4.Standard deviation
# √291.67 = 17.08

print(np.std(arr))
```

#Indexing and slicing


```python
# indexing - accessing a specific element using its position

#1D

arr = np.array([1,2,3,4,5])
print(arr[0])
print(arr[2])
print(arr[-1])

#2D

arr = np.array([[1,2,3],[4,5,6]])
print(arr[0,1])
print(arr[1,2])
```

```python
#slicing - getting a range of elements

#1D
arr = np.array([10,20,30,40,50])
print(arr[1:4])
print(arr[:3])
print(arr[::2])

#2D
arr = np.array([[1,2,3],[4,5,6]])
print(arr[:,1])
print(arr[0,:])
print(arr[0:2,1:3])
```

```python
# Boolean Indexing - filtering elements using condition

#1D
arr = np.array([10,20,30,40,50])
print(arr[arr>25])

arr = np.array([1,2,3,4,5,6])
print(arr[arr%2==0])

#2D
arr = np.array([[1,2,3],[4,5,6]])
print(arr[arr>3])
```
