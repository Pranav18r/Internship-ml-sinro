### class distribution - handling unbalanced data distribution

handling imbalanced data , when we do classification , we need variety of data , but we observe a majority of data available in one specific area causing imbalance in training , we have to correct this , for this we do undersampling

checking fraudulent and legitimate transactions in credit card transactions

```python
import numpy as np
import pandas as pd
```

```python
#seperating legit and fraudulent transaction

legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
legit.shape
```

```python
fraud.shape
```

undersampling , building a sampple dataset containing distribution of legit and fraudulent transaction

no of fraudulent transaction = 492

```python
legit_sample = legit.sample(n = 492) #we reduced the no of samples of legitimate
```

```python
legit_sample.shape
```

```python
type(legit_sample)
```

```python
#concatinate the 2 dataframes

new_dataset = pd.concat([legit_sample,fraud],axis = 0)
```
