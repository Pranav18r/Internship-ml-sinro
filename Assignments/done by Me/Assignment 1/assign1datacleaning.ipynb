Step 2.2 - Checking for irregular , or data which is not true , 

for eg : age can never be 2000

````python
#setting constraints
constraints = {
    'age': (0, 120),
    'experience_yrs': (0, 60),
    'education_yrs': (0, 30),
    'hours_per_week': (0, 168),
    'iq_score': (40, 200),
    'city_population': (0, 1_000_000_000),
    'distance_to_office_km': (0, 1000),
    'avg_temp_celsius': (-60, 60),
    'prev_salary': (0, 1_000_000_000),
    'bonus_pct': (0, 100),
    'annual_salary': (0, 1_000_000_000)
}
````

````python
#to view the values in which column violate the constraint
for column, (min_val, max_val) in constraints.items():

    invalid_values = read_data[
        (read_data[column] < min_val) |
        (read_data[column] > max_val)
    ]

    if not invalid_values.empty:
        print(f"\nColumn: {column}")
        print(invalid_values[column])
````

````python
#replace the cells that violate the constraint with nan
for column, (min_val, max_val) in constraints.items():

    read_data[column] = read_data[column].mask(
        (read_data[column] < min_val) |
        (read_data[column] > max_val),
        np.nan
    )
````

````python
#now check if its managed
for column, (min_val, max_val) in constraints.items():

    invalid_values = read_data[
        (read_data[column] < min_val) |
        (read_data[column] > max_val)
    ]

    if not invalid_values.empty:
        print(f"\nColumn: {column}")
        print(invalid_values[column])
````

````python
#replace missing values with central tendencies 

# Replacing missing values in continuous numeric columns using median to reduce outlier impact.
# Replacing missing values in categorical/code-based columns using mode to preserve most frequent category.

continuous_columns = [
    'age',
    'experience_yrs',
    'education_yrs',
    'hours_per_week',
    'iq_score',
    'city_population',
    'distance_to_office_km',
    'avg_temp_celsius',
    'prev_salary',
    'bonus_pct',
    'annual_salary'
]

categorical_columns = [
    'dept_code',
    'gender_code'
]

# Fill continuous columns with median
for column in continuous_columns:
    read_data[column] = read_data[column].fillna(
        read_data[column].median()
    )

# Fill categorical columns with mode
for column in categorical_columns:
    read_data[column] = read_data[column].fillna(
        read_data[column].mode()[0]
    )
````

# Handling Missing Values by Data Type

## 1. Numeric Columns

Use central tendencies.

### Median (best for skewed/outlier-heavy data)

```python
df['salary'] = df['salary'].fillna(
    df['salary'].median()
)
```

### Mean (best for normal distributions)

```python
df['age'] = df['age'].fillna(
    df['age'].mean()
)
```

---

## 2. Categorical / String Columns

### Fill with placeholder

```python
df['city'] = df['city'].fillna('Unknown')
```

### Fill with mode (most common value)

```python
df['city'] = df['city'].fillna(
    df['city'].mode()[0]
)
```

### Leave missing

Useful when missing itself has meaning.

```python
# keep NaN as-is
```

Examples:

* missing middle name
* missing comments

---

## 3. Date Columns

### Forward Fill

Uses previous valid value.

```python
df['date'] = df['date'].fillna(method='ffill')
```

Useful for:

* time series
* stock data
* sensor logs

### Backward Fill

```python
df['date'] = df['date'].fillna(method='bfill')
```

### Interpolation

```python
df['date'] = df['date'].interpolate()
```

### Placeholder Date

```python
df['date'] = df['date'].fillna(
    pd.Timestamp('1900-01-01')
)
```

---

## 4. ID Columns

IDs should usually remain unique.

### Remove rows with missing IDs

```python
df = df.dropna(subset=['employee_id'])
```

### Never use:

* mean
* median

Examples:

* employee_id
* transaction_id

---

## 5. Boolean Columns

### Fill with False

```python
df['is_active'] = df['is_active'].fillna(False)
```

### Or use mode

```python
df['is_active'] = df['is_active'].fillna(
    df['is_active'].mode()[0]
)
```

---

# General Rule

| Data Type   | Common Strategy             |
| ----------- | --------------------------- |
| Numeric     | Mean / Median               |
| Categorical | Mode                        |
| Text        | Unknown / Keep NaN          |
| Dates       | ffill / bfill / interpolate |
| IDs         | Drop rows                   |
| Boolean     | False / Mode                |
