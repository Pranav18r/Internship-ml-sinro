## seaborn countplot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# NumPy array
arr = np.array(["A", "B", "A", "C", "B", "A", "C", "C", "B"])

# Convert to Pandas Series
series = pd.Series(arr)

# Create countplot
sns.countplot(x=series)

# Display plot
plt.title("Countplot from NumPy Array and Pandas Series")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    "Class": ["A", "A", "B", "B", "C", "C"],
    "Gender": ["Boys", "Girls", "Boys", "Girls", "Boys", "Girls"],
    "Average Attendance": [85, 90, 80, 88, 82, 91]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create barplot with hue
sns.barplot(
    x="Class",
    y="Average Attendance",
    hue="Gender",
    data=df
)

# Add title
plt.title("Boys vs Girls Average Attendance")

# Show plot
plt.show()
```

## seaborn histogram

```python
# plot histogram of salary distribution with KDE
import seaborn as sns
import matplotlib.pyplot as plt

salary = [25000, 30000, 28000, 35000, 40000,
          42000, 50000, 52000, 60000, 62000]

sns.histplot(salary, kde=True)

plt.title("Salary Distribution with KDE")
plt.xlabel("Salary")
plt.ylabel("Count")

plt.show()
```

```python
# 3.Generate random numbers using Numpy and visualize distribution.
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)

plt.hist(data, bins=30)

plt.title("Random Number Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Math": [90,85,78,92,88,76,95,89,84,73],
    "Science": [88,80,75,96,90,70,98,85,82,72],
    "English": [75,82,70,91,86,74,93,80,79,68],
    "StudyHours": [5,4,3,6,5,2,7,5,4,2],
    "Attendance": [95,90,85,98,96,80,99,94,88,78],
})

corr = df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)

plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female"],
    "Math": [90, 78, 85, 92, 70, 88],
    "Science": [88, 82, 80, 95, 72, 90],
    "English": [75, 85, 70, 90, 68, 92],
    "Attendance": [95, 90, 85, 98, 80, 96],
    "StudyHours": [5, 4, 3, 6, 2, 5]
})

plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
sns.countplot(x="Gender", data=df)
plt.title("Gender Count")

avg_marks = df[["Math","Science","English"]].mean()

plt.subplot(2,2,2)
sns.barplot(x=avg_marks.index, y=avg_marks.values)
plt.title("Average Marks Comparison")

plt.subplot(2,2,3)
sns.histplot(df["Attendance"], bins=5, kde=True)
plt.title("Attendance Distribution")

plt.subplot(2,2,4)
sns.scatterplot(x="StudyHours", y="Math", hue="Gender", data=df)
plt.title("Study Hours vs Math Marks")

plt.tight_layout()
plt.show()
```
