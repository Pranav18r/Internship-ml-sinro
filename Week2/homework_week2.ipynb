#### 1.  Encoding ( label & one hot)

```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

df = pd.read_csv("/home/silvy_linux/Desktop/Internship/Datasets/Titanic-Dataset.csv")
print(df)
```

```python
# Label Encoding 

df['Survived'].value_counts()

label_encode = LabelEncoder()

labels = label_encode.fit_transform(df['Survived'])
print(labels)

df['survived_labels'] = labels
print(df.head())
```

```python
# One Hot Encoding 

ohe = OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')

ohetransform = ohe.fit_transform(df[["Sex"]])
print(ohetransform)

df = pd.concat([df,ohetransform],axis='columns')
df.drop('Sex',axis='columns',inplace=True)
print(df)
```

#### 2. Standarisation

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Loading dataset
df = pd.read_csv("/home/silvy_linux/Desktop/Internship/Datasets/Titanic-Dataset.csv")

print(df.head()) 

# Keeping only numerical columns
df = df.select_dtypes(include=np.number)

# Input and Output
x = df.drop("Survived", axis=1)
y = df["Survived"]
```

```python
print(x)
print(y)
```

```python
# Splitting dataset
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=3
)

print(x.shape)
print(x_train.shape)
print(x_test.shape)

# Standard deviation before standardisation
print("Before Standardisation")
print(x.std())

# Standardisation
scaler = StandardScaler()
```

```python
scaler.fit(x_train)

x_train_scaler = scaler.transform(x_train)
x_test_scaler = scaler.transform(x_test)

# Converting into DataFrame
x_train_scaler = pd.DataFrame(x_train_scaler, columns=x.columns)
x_test_scaler = pd.DataFrame(x_test_scaler, columns=x.columns)

print(x_train_scaler)
print(x_test_scaler)

# Checking SD after standardisation
print("Train Dataset SD")
print(x_train_scaler.std())

print("Test Dataset SD")
print(x_test_scaler.std())
```

#### 3. Normalisation
