```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Data Initialization
df = pd.DataFrame({
    "Name": ["A","B","C","D","E"],
    "Math": [90,70,85,60,95],
    "Science": [88,65,80,72,98],
    "English": [75,85,78,80,92],
    "Attendance": [95,88,90,85,98],
    "Study Hours": [5,3,4,2,6]
})

# 2. Subplot Configuration
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Comprehensive Student Performance Analysis', fontsize=20, fontweight='bold')
subjects = ["Math", "Science", "English"]

# --- CHART 1: LINE PLOT ---
student_a = df[df["Name"] == "A"][subjects].iloc[0]
axes[0, 0].plot(subjects, student_a, marker='o', color='dodgerblue')
axes[0, 0].set_title("1. Subject Marks (Student A)")
axes[0, 0].set_xlabel("Academic Subjects")
axes[0, 0].set_ylabel("Marks (Out of 100)")

# --- CHART 2: BAR CHART ---
avg_marks = df[subjects].mean()
axes[0, 1].bar(subjects, avg_marks, color=['#ff9999','#66b3ff','#99ff99'], edgecolor='black')
axes[0, 1].set_title("2. Average Marks across Subjects")
axes[0, 1].set_xlabel("Academic Subjects")
axes[0, 1].set_ylabel("Average Score")

# --- CHART 3: SCATTER PLOT ---
axes[0, 2].scatter(df["Study Hours"], df["Math"], color='darkorchid', s=120)
axes[0, 2].set_title("3. Correlation: Study Time vs. Math Result")
axes[0, 2].set_xlabel("Weekly Study Hours")
axes[0, 2].set_ylabel("Math Examination Mark")

# --- CHART 4: HISTOGRAM ---
axes[1, 0].hist(df["Attendance"], bins=5, color='gold', edgecolor='black')
axes[1, 0].set_title("4. Frequency of Attendance Levels")
axes[1, 0].set_xlabel("Attendance Percentage Range (%)")
axes[1, 0].set_ylabel("Number of Students (Frequency)")

# --- CHART 5: PIE CHART ---
total_marks = df[subjects].sum()
axes[1, 1].pie(total_marks, labels=subjects, autopct='%1.1f%%', startangle=140)
axes[1, 1].set_title("5. Contribution to Total Group Score")

# Remove unused axes
axes[1, 2].axis('off') 

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
