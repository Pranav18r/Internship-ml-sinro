## using onehot encoder

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
path = "/home/silvy_linux/Desktop/Internship/Datasets/housing_data.csv"
df = pd.read_csv(path)
```

```python
df
```

```python
ohe = OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
```

```python
ohetransform = ohe.fit_transform(df[['town']])
```

```python
ohetransform
```

```python
df = pd.concat([df,ohetransform],axis='columns')
```

```python
df
```

```python
df.drop(['town'],axis='columns',inplace=True)
```

```python
df
```

## performing encoding techniques using this

practise of both - label and one hot encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
d = {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000],
'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'],
'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small',]}

df = pd.DataFrame(d)
```

```python
print(df)
```

```python
#performing one hot encoding on city
ohe = OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
```

```python
ohetransform = ohe.fit_transform(df[['city']])
```

```python
ohetransform
```

```python
df = pd.concat([df,ohetransform],axis='columns')
```

```python
df.drop(['city'],axis='columns',inplace=True)
```

```python
df
```

```python
#now performing label encoding
df["size"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(df['size'])
```

```python
labels
```

```python
df.head()
```

```python
df_new = df.drop("size",axis = 1)
```
