## Standardization
 is a crucial preprocessing technique in machine learning that rescales numerical features to have a mean of zero (\(\mu = 0\)) and a standard deviation of one (\(\sigma = 1\))

```python
import numpy as numpy
import pandas as pd
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#loading the dataset

dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
df = pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
df
```

```python
#imput and output
x = df
y = dataset.target
```

```python
x
```

```python
y
```

```python
#splitting the data into training data and test dataset

x_train , x_test , y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
print(x_train,x_test)
```

```python
print(y_train,y_test) #its already in 0 and 1 , therefore you don't need to standardize
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#standardising the data
```

```python
print(dataset.data.std())
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train) #analayse and understand how data is distributed in x_train
```

```python
#transform the data based on StandardScaler()
```

```python
x_train_standardized = scaler.transform(x_train)
print(x_train_standardized)
```

```python
x_test_standardized = scaler.transform(x_test)
print(x_test_standardized)
```

```python
print(x_train_standardized.std())
```

```python
print(x_test_standardized.std())
```

Normalization and standardization are feature scaling techniques used to prepare data for machine learning, with normalization scaling data to a fixed range (usually [0, 1]) and standardization scaling data to a mean of 0 and standard deviation of 1. Normalization is ideal for distance-based algorithms like KNN, while standardization works better for algorithms assuming normal
