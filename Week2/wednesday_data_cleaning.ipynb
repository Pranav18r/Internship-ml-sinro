```python
import pandas as pd

path = "/home/silvy_linux/Desktop/Internship/Datasets/workout_data.csv"

df = pd.read_csv(path)

# Fill missing Calories with mode
mo = df["Calories"].mode()[0]
df.fillna({"Calories": mo}, inplace=True)

# Convert Date column
df["Date"] = pd.to_datetime(df["Date"], format="mixed")

# Remove invalid dates
df.dropna(subset=["Date"], inplace=True)

# Remove rows where Duration > 120
for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.drop(x, inplace=True)

print(df)
```

```python
print(df.duplicated())
```
