```python
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Department": ["HR", "IT", "HR", "IT", "HR"],
    "Salary": [30000, 50000, 35000, 60000, 40000]
}

df = pd.DataFrame(data)

# 1. Filter employees with Salary > 40000
filtered_df = df[df["Salary"] > 40000]
print("Filtered Employees:")
print(filtered_df)

# 2. Sort dataset by Salary (descending)
sorted_df = df.sort_values(by="Salary", ascending=False)
print("\nSorted Dataset:")
print(sorted_df)

# 3. Find average salary per department
avg_salary = df.groupby("Department")["Salary"].mean()
print("\nAverage Salary per Department:")
print(avg_salary)

# 4. Find total salary per department
total_salary = df.groupby("Department")["Salary"].sum()
print("\nTotal Salary per Department:")
print(total_salary)

# 5. Find highest salary in each department
highest_salary = df.groupby("Department")["Salary"].max()
print("\nHighest Salary in Each Department:")
print(highest_salary)
```
