```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

marks = np.array([10,20,15,30,50])
plt.plot(marks)
plt.xlabel("Marks")
plt.ylabel("Performance index")
plt.show()
```

```python
temperature = pd.Series([30,32,35,33,31,29])
plt.plot(temperature)
plt.xlabel("temperature")
plt.show()
```

```python
marks = np.array([10,20,15,30,50])
plt.plot(marks)
marks2 = np.array([30,40,50,10,20])
plt.plot(marks2)
plt.legend("marks_no_1")
plt.legend("marks_no_2")
plt.show()
```
